# ⚙️ AI System Variables and Their Units

| Variable | Unit     | Description of Unitization |
|----------|----------|-----------------------------|
| `E` (Energy)        | `cUnit`  | 1 `cUnit` = Compute needed to process 1 token through 1 transformer layer on 1 GPU. Represents raw computational fuel (FLOPs, power, latency). |
| `C` (Cognition)     | `Cog`    | 1 `Cog` = Ability to model, reason, and simulate 1 distinct abstract concept or logical inference. Represents internal reasoning capacity. |
| `M` (Memory)        | `Mem`    | 1 `Mem` = 1 retrievable and usable memory item (fact, token span, or pattern). Can be short-term (`Mem_s`) or long-term (`Mem_l`). |
| `A` (Actionability) | `Act`    | 1 `Act` = 1 successful external operation (tool call, decision, task execution). Represents the agent's capacity to affect the world. |
| `G` (Goal Alignment)| `Align`  | 1 `Align` = Degree of consistency, clarity, and prioritization of a single internal goal over time. Higher values mean more coherent behavior. |
| `K` (Knowledge)     | `Know`   | 1 `Know` = 1 verified, applicable unit of information. Represents learned or pretrained concepts available for reasoning or retrieval. |
| `S` (Self-Reflection)| `Meta`  | 1 `Meta` = Capacity to observe, evaluate, and improve 1 internal cognitive or behavioral loop. Enables introspection, error correction, and learning. |
